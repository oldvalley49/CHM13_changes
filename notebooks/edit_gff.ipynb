```python
def remove_feature(df, row):
    next_df = df[row.name + 1:]
    next_gene = next_df[next_df['type'] == 'gene']
    if next_gene.empty:
        drop_idx = df.loc[row.name : ].index
    else:
        drop_idx = df[row.name:next_gene.iloc[0].name].index
    return df.drop(drop_idx)
```

```python
test = remove_feature(JHU_df, match)
```

```python
def remove_genes(df, remove_df):
    for _, gene in remove_df.iterrows():
        matches = df[
            (df.type == 'gene')
            & (df.gene == gene.gene)
            & (df.seq_id == gene.chromosome)
            & (df.start == gene.start)
            & (df.end == gene.end)
            & (df.strand == gene.strand)
        ]
        if matches.empty:
            continue
        else:
            match = matches.iloc[0]
            df = remove_feature(df, match)
    return df
```

```python
test = remove_genes(JHU_df, remove)
```
